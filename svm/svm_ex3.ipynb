```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
c1 = np.array([[2,2],[3,3],[4,4],[5,5],[4,6],[3,7],[4,8],[5,9],[6,10]])
c2 = np.array([[6,2],[7,3],[8,4],[9,5],[8,6],[7,7],[7,8],[7,9],[8,10]])
```

```python
u1 = np.mean(c1,axis=0)
u2 = np.mean(c2,axis=0)
```

```python
plt.figure(figsize = (10,10))
plt.grid()
plt.scatter(c1.T[0],c1.T[1],label = 'c1')
plt.scatter(c2.T[0],c2.T[1],label = 'c2')
plt.scatter(u1.T[0],u1.T[1],marker='_', s=200,linewidth=4)
plt.scatter(u2.T[0],u2.T[1],marker='+', s=200,linewidth=4)
plt.legend()
plt.show()
```
